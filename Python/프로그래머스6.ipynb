## 삼총사

```python
def solution(number):
    answer = 0 # 삼총사를 만들 수 있는 방법의 수
    n = len(number)
    
    for i in range(0, n-2): # 첫번째 정수
        for j in range(i+1, n-1): # 두번째 정수
            for k in range(j+1, n): # 세번째 정수 
                if number[i] + number[j] + number[k] == 0:
                    answer += 1
    return answer

print(solution([-2, 3, 0, 2, -5])) # 2
print(solution([-3, -2, -1, 0, 1, 2, 3])) # 5
print(solution([-1, 1, -1, 1])) # 0
```

## 시소 짝꿍

다 몸무게가 같으면? 조합 공식  
같은 몸무게일 때랑 다른 몸무게일 때로 나눠보기


```python
def solution(weights): 
    answer = 0 # 시소짝꿍 수(몇 쌍인지)
    freq = {} # 몸무게(key):등장횟수(value)
    for w in weights:
        if w in freq: # 이미 있는 몸무게면 
            freq[w] += 1 # 등장횟수 누적 
        else: # 새로운 몸무게이면 
            freq[w] = 1 # 등장횟수 1
    print(freq)

    for w in freq: # 몸무게를 하나씩 확인
        n = freq[w] # 몸무게가 w인 사람 수 

        answer += n * (n - 1) // 2 * 1 # 같은 몸무게일 때 짝꿍 조합

        if (w * 3) % 2 == 0: # 3:2(정수 조건)
            p = w * 3 / 2
            if p in freq:
                answer += n * freq[p] # 몸무게가 w인 사람수 * 몸무게가 p인 사람수

        if (w * 4) % 2 == 0: # 4:2(정수 조건)
            p = w * 4 / 2
            if p in freq:
                answer += n * freq[p]

        if (w * 4) % 3 == 0: # 4:3(정수 조건)
            p = w * 4 / 3
            if p in freq:
                answer += n * freq[p]
    return answer

print(solution([100, 180, 360, 100, 270]))  # 4
```
